```python
your_key = "zgiP0TzXASkJJPUQ9BT7AdL8GEcnOA5k"
your_endpoint = "c5sfbduq"
import json
import urllib
import pandas as pd

def getBestPicture(api_key=your_key,endpoint=your_endpoint):
    """
    Creates list of Best Picture Oscar winners by year,
    plus their length and IMDB rating.
    """
    
    years, movies, lengths, ratings = [], [], [], []
    #Remember to replace this link with the link to your specifc API
    url = "https://www.kimonolabs.com/api/{}?".format(endpoint) + \
            "apikey={}".format(api_key)
    data = json.load(urllib.urlopen(url))
    
    # Iterate through json object to collect data
    for n in xrange(data['count']):
        n_year = int(data['results']['collection1'][n]['Year'][1:-1])
        n_title = data['results']['collection1'][n]['Film Name']['text']
        n_length = float(data['results']['collection1'][n]['Length'][0:-6])
        n_rating = float(data['results']['collection1'][n]['Rating'])
        years.append(n_year)
        movies.append(n_title)
        lengths.append(n_length)
        ratings.append(n_rating)
    data = pd.DataFrame({'movie':movies,'rating':ratings,'length_in_min':lengths,'year':years})
    
    return data
movies = getBestPicture()
movies.head(50)
```
